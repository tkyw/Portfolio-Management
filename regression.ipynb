```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_excel(r"C:\Users\USER\Python Programme\uni\portfolio_management\regression.xlsx", index_col=0)
df
```

```python
market_index = np.array(df['S_P_500'])
stocks = df.iloc[:, 1:]
items = {}
slopes = []
intercepts = []
expected_returns = []
stds = []
# fit a line to the data
for ticker in stocks.columns:
    stock = np.array(df[ticker])
    slope, intercept = np.polyfit(market_index, stock, 1)
    slopes.append(slope)
    intercepts.append(intercept)
    expected_returns.append(df[ticker].mean())
    stds.append(df[ticker].std(ddof=1))
# print("Slope:", slope)
items['beta'] = slopes
items['intercepts'] = intercepts
items['tickers'] = stocks.columns
items['Expected Return'] = expected_returns
items['Standard Deviation'] = stds
df_slopes = pd.DataFrame(items)
df_slopes = df_slopes.rename(columns= {
    0: "beta"
}).sort_values(by="beta",ascending=True).set_index("tickers", drop=True)
```
