```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv("return.csv", index_col=0)
df = df.replace("%", "", regex=True)
df = df.astype(float)
bench = df.iloc[:, -2]
df = df.iloc[:, :-2]
```

```python
bench
```

```python
excess_return = df.sub(bench, axis=0)
average = excess_return.mean()
std = excess_return.std(ddof=1)
IR = average / std
IR.to_clipboard(excel=True)
average
```

```python
semi_std = []
for col in df.columns:
    data = df[col]
    downside = data[(data - data.mean()) < 0]
    std = ((((downside - data.mean()) ** 2).sum() / (len(downside) -1 )) ** 0.5) / 100
    semi_std.append(std)
```

### Semi STD of S&P

```python
bench_downside = bench[(bench - bench.mean()) < 0]
std = ((((bench_downside - bench.mean()) ** 2).sum() / (len(bench_downside) -1 )) ** 0.5) / 100
std
```

```python
df_sd = pd.DataFrame(semi_std, index=df.columns, columns=['Semi Std'])
df_sd.to_clipboard(excel=True, index=False)
```
